```python
# Challenge 1 - Most Profiting Authors

"""
1. Calculate the royalty of each sale for each author.

2. Using the output from Step 1 as a sub-table, aggregate the total royalties for each title for each author.

3. Using the output from Step 2 as a sub-table, 
calculate the total profits of each author by aggregating the advances and total royalties of each title.
"""
```

```python
#Challenge 1 Step 1 - Calculate the royalty of each sale for each author.

"""
Write a SELECT query to obtain the following output:

Title ID
Author ID
Royalty of each sale for each author

The formula is:
sales_royalty = titles.price * sales.qty * titles.royalty / 100 * titleauthor.royaltyper / 100

Note that titles.royalty and titleauthor.royaltyper are divided by 100 respectively because they are percentage numbers 
instead of floats.

In the output of this step, each title may appear more than once for each author. 
This is because a title can have more than one sales.
"""
```

```python
royalty_per_sale = """
SELECT 
    T.title_id AS TITLE_ID,
    A.au_id AS AUTHOR_ID,
    A.au_lname AS NAME,
    T.title AS TITLE,
    TA.royaltyper AS ROYALTY_PER,
    T.advance AS ADVANCE,
    T.price,
    S.qty,
    T.royalty,
    T.price * S.qty * T.royalty / 100 * TA.royaltyper / 100 AS SALES_ROYALTY
FROM
    `ironhack-data-analytics-265219.publications.authors` AS A
LEFT JOIN 
    `ironhack-data-analytics-265219.publications.titleauthor` AS TA
ON 
    A.au_id=TA.au_id
LEFT JOIN
    `ironhack-data-analytics-265219.publications.titles` AS T
ON
    TA.title_id=T.title_id
LEFT JOIN
    `ironhack-data-analytics-265219.publications.sales` AS S
ON
    T.title_id=S.title_id
WHERE 
    T.title_id IS NOT NULL
GROUP BY
    T.title_id,
    A.au_id,
    A.au_lname,
    T.title,
    T.advance,
    TA.royaltyper,
    T.price,
    S.qty,
    T.royalty
ORDER BY
    TITLE_ID;
"""
```

```python
query_job = client.query(query=royalty_per_sale)
```

```python
query_job.to_dataframe()
```

```python
#Challenge 1 Step 2 - Aggregate the total royalties for each title for each author

"""
Using the output from Step 1, write a query to obtain the following output:

Title ID
Author ID
Aggregated royalties of each title for each author

Hint: use the SUM subquery and group by both au_id and title_id

In the output of this step, each title should appear only once for each author.
"""
```

```python
royalty_per_t_per_a = """
SELECT
    TITLE_ID,
    AUTHOR_ID,
    NAME,
    (ADVANCE * R.ROYALTY_PER/100) AS TOTAL_ADVANCE,
    ROYALTY_PER,
    SUM(R.SALES_ROYALTY) AS ROYALTY_PER_T_PER_A            
FROM
    (SELECT
        T.title_id AS TITLE_ID,
        A.au_id AS AUTHOR_ID,
        A.au_lname AS NAME,
        T.title AS TITLE,
        TA.royaltyper AS ROYALTY_PER,
        T.advance AS ADVANCE,
        T.price,
        S.qty,
        T.royalty,
        (T.price * S.qty * T.royalty / 100 * TA.royaltyper / 100) AS SALES_ROYALTY
    FROM
        `ironhack-data-analytics-265219.publications.authors` AS A
    LEFT JOIN 
        `ironhack-data-analytics-265219.publications.titleauthor` AS TA
    ON 
        A.au_id=TA.au_id
    LEFT JOIN
        `ironhack-data-analytics-265219.publications.titles` AS T
    ON
        TA.title_id=T.title_id
    LEFT JOIN
        `ironhack-data-analytics-265219.publications.sales` AS S
    ON
        T.title_id=S.title_id
    WHERE 
        T.title_id IS NOT NULL
    GROUP BY
        T.title_id,
        A.au_id,
        A.au_lname,
        T.title,
        T.advance,
        TA.royaltyper,
        T.price,
        S.qty,
        T.royalty
    ORDER BY
        TITLE_ID) AS R
GROUP BY
    TITLE_ID,
    AUTHOR_ID,
    NAME,
    ADVANCE,
    ROYALTY_PER
ORDER BY
    AUTHOR_ID;
"""
```

```python
query_job = client.query(query=royalty_per_t_per_a)
```

```python
query_job.to_dataframe()
```

```python
#Challenge 1 Step 3 - Calculate the total profits of each author

"""
Now that each title has exactly one row for each author where the advance and royalties are available, 
we are ready to obtain the eventual output. 

Using the output from Step 2, write a query to obtain the following output:

Author ID
Profits of each author by aggregating the advance and total royalties of each title

Sort the output based on a total profits from high to low, and limit the number of rows to 3.
"""
```

```python
total_profit = """
SELECT
    AUTHOR_ID,
    NAME,
    SUM(royalties_per_a.ROYALTY_PER_T_PER_A + royalties_per_a.TOTAL_ADVANCE) AS PROFIT
FROM
    (SELECT
        TITLE_ID,
        AUTHOR_ID,
        NAME,
        (ADVANCE * royalties_per_sale.ROYALTY_PER/100) AS TOTAL_ADVANCE,
        ROYALTY_PER,
        SUM(royalties_per_sale.SALES_ROYALTY) AS ROYALTY_PER_T_PER_A            
    FROM
        (SELECT
            T.title_id AS TITLE_ID,
            A.au_id AS AUTHOR_ID,
            A.au_lname AS NAME,
            T.title AS TITLE,
            TA.royaltyper AS ROYALTY_PER,
            T.advance AS ADVANCE,
            T.price,
            S.qty,
            T.royalty,
            (T.price * S.qty * T.royalty / 100 * TA.royaltyper / 100) AS SALES_ROYALTY
        FROM
            `ironhack-data-analytics-265219.publications.authors` AS A
        LEFT JOIN 
            `ironhack-data-analytics-265219.publications.titleauthor` AS TA
        ON 
            A.au_id=TA.au_id
        LEFT JOIN
            `ironhack-data-analytics-265219.publications.titles` AS T
        ON
            TA.title_id=T.title_id
        LEFT JOIN
            `ironhack-data-analytics-265219.publications.sales` AS S
        ON
            T.title_id=S.title_id
        WHERE 
            T.title_id IS NOT NULL
        GROUP BY
            T.title_id,
            A.au_id,
            A.au_lname,
            T.title,
            T.advance,
            TA.royaltyper,
            T.price,
            S.qty,
            T.royalty
        ORDER BY
            TITLE_ID) AS royalties_per_sale
    GROUP BY
        TITLE_ID,
        AUTHOR_ID,
        NAME,
        ADVANCE,
        ROYALTY_PER) AS royalties_per_a
GROUP BY
    AUTHOR_ID,
    NAME
ORDER BY
    PROFIT DESC
LIMIT 3
"""
```

```python
query_job = client.query(query=total_profit)
```

```python
query_job.to_dataframe()
```

```python
# Challenge 2

"""
Elevating from your solution in Challenge 1 , create a table named most_profiting_authors to hold the data about 
the most profiting authors. The table should have 2 columns:

au_id - Author ID
profits - The profits of the author aggregating the advances and royalties
"""
```

```python
sql = """
CREATE TABLE publications.most_profiting_authors_kristina AS
    SELECT
        AUTHOR_ID,
        NAME,
        SUM(royalties_per_a.ROYALTY_PER_T_PER_A) + SUM(royalties_per_a.ADVANCE*royalties_per_a.ROYALTY_PER/100) AS PROFIT
    FROM
        (SELECT
            TITLE_ID,
            AUTHOR_ID,
            NAME,
            ADVANCE,
            ROYALTY_PER,
            SUM(royalties_per_sale.SALES_ROYALTY) AS ROYALTY_PER_T_PER_A            
        FROM
            (SELECT
                T.title_id AS TITLE_ID,
                A.au_id AS AUTHOR_ID,
                A.au_lname AS NAME,
                T.title AS TITLE,
                TA.royaltyper AS ROYALTY_PER,
                T.advance AS ADVANCE,
                T.price,
                S.qty,
                T.royalty,
                (T.price * S.qty * T.royalty / 100 * TA.royaltyper / 100) AS SALES_ROYALTY
            FROM
                `ironhack-data-analytics-265219.publications.authors` AS A
            LEFT JOIN 
                `ironhack-data-analytics-265219.publications.titleauthor` AS TA
            ON 
                A.au_id=TA.au_id
            LEFT JOIN
                `ironhack-data-analytics-265219.publications.titles` AS T
            ON
                TA.title_id=T.title_id
            LEFT JOIN
                `ironhack-data-analytics-265219.publications.sales` AS S
            ON
                T.title_id=S.title_id
            WHERE 
                T.title_id IS NOT NULL
            GROUP BY
                T.title_id,
                A.au_id,
                A.au_lname,
                T.title,
                T.advance,
                TA.royaltyper,
                T.price,
                S.qty,
                T.royalty
            ORDER BY
                TITLE_ID) AS royalties_per_sale
        GROUP BY
            TITLE_ID,
            AUTHOR_ID,
            NAME,
            ADVANCE,
            ROYALTY_PER) AS royalties_per_a
    GROUP BY
        AUTHOR_ID,
        NAME
    ORDER BY
        PROFIT DESC
    LIMIT 3
"""
```

```python
query_job = client.query(query=sql)
```

```python
query_job.to_dataframe()
```
